```python
user_books2 = ds.pivot_table(index='ID',
                        columns='Book',
                        values='Rating').reset_index(drop=True)
```

```python
user_books2
```

```python
user_books2.index = ds.ID.unique()
```

```python
user_books2
```

```python
user_books2.fillna(0, inplace=True)
```

```python
user_books2
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_books2.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```
